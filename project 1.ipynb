#### 1.Display the count of matches based on the year wise
#### Description of the problem
   - We read the csv file and convert it to a dataframe
   - Then we use groupby clause and get the count of matches based on year

```python
import pandas as pd
filepath='DataFile\Match.csv'
def readCsvData(filepath):
    return pd.read_csv(filepath)
df=readCsvData(filepath)
ad=pd.DataFrame(df.groupby('Season_Year')['Match_SK'].count())
ad
```

### 2. Plot a bar graph no. of matches won by each team in the year 2011
#### Description Of Program
  - We need import matplotlib and numpy packages as we need to produce graphs
  - We take a sub Data Base and store it diferently seperated under condition year 2011
  - We take the teams participated and number of wins for each team
  - With the above data we create bar graph

```python
import matplotlib.pyplot as plt
import numpy as np
df2=df.copy()
sdf=pd.DataFrame(df2[df2['Season_Year'].isin(['2011'])])
wins=(sdf.groupby('match_winner')['Match_SK'].count())
teams=sdf.Team1.unique()
index=np.arange(len(teams))
plt.barh(teams,wins,color='red')
```

#### 3. Print ManOfMatch count of each player in Hyderabad stadium
#### Description of the problem
  - Seperating the data base to a sub data base divided by city name 'Hyderbad' using isin()
  - Grouping the man of match using groupby clause and using count() gives man of match count

```python
df2=df.copy()
sdf=pd.DataFrame(df2[df2['City_Name'].isin(['Hyderabad'])])
pd.DataFrame(sdf.groupby('ManOfMach')['Match_SK'].count())
```
