```python
def sum(n,s):
    s=s+n
    return s
sum1=0
r=0
x=int(input())
while x!=0:
    r=x%10
    if r%2==0:
        sum1=sum(r,sum1)
    x=x//10
print(sum1)
```

```python
# input:19535
# output : 9(largest digit of the given number)
# Functional programming
def largestDigit(n):
    l=n%10
    n=n//10
    while n!=0:
        r=n%10
        if l<r:
            l=r
        n=n//10
    return l
largestDigit(int(input()))
```

```python
#Read a number as input 
# output is yes or no
# input : 145
# output : yes
#input : 123
# output : no
# individual factorial sum of digits
def fact(n):
    f=1
    for i in range (1,n+1):
         f=f*i
    return f
def sum(x):
    s=0
    n=x
    while x!=0:
        r=x%10
        s=s+fact(r)
        x=x//10
    if n==s:
        return 'Yes'
    else : 
        return 'No'
   
sum(int(input()))
```

```python
# Read a number and print if its a palindrome
def palin(n):
    x=n
    s=0
    while n!=0:
        r=n%10
        s=s*10+r
        n=n//10
    if x==s:
        return "Number is palindrome"
    else: return "Number is not palindrome"
palin(int(input()))
```

```python
# Print numbers in between given range
def printseries(lb,ub):
    for i in range (lb,ub+1):
        print(i,end=' ')
printseries(int(input()),int(input()))
```

```python
# funtion to print the alternative numbers

def series(lb,ub):
    for i in range (lb,ub+1,2):
        print(i,end=',')
series(int(input()),int(input()))
```

```python
# Print the prime numbers between given m ana n

def prime(a):
    c=0
    for i in range(2,a):
        if a%i==0:
            c=1
            break;
    if c==0:
        print(a,end=' ')
def sol(m,n):
    print("Prime Numbers")
    for i in range (m,n):
        prime(i)
sol(int(input()),int(input()))
```
