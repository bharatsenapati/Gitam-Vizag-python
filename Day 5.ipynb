### Idiomatic programming
 - Transformation code into beatiful,idiomatic python
 - Replaces traditional index manipulation with python code looping idoms

```python
# looping over a range of number
for i in [0,1,2,3,4,5]:
    print(i**2,end=' ')
print()
# pythinic way
for i in range(6):
    print(i**2,end=' ')
```

```python
# Looping over the collection(List)
li=[1,2,3,4,5]
for i in range(len(li)):
    print(li[i],end=' ')
print()
# Pythonic way
for i in li:
    print(i,end=' ')
```

```python
# looping from Backwards
li=[1,2,3,4,5]
for i in range(len(li)-1,-1,-1):
    print(li[i],end=" ")
# pythonic Way
print()
for i in reversed(li):
    print(i,end=' ')
```

```python
# Looping over the collection with index
li=[1,2,3,4,5]
for i in range(len(li)):
    print(i,"-->",li[i])
print()
# Pythonic way
for i,a in enumerate(li):
    print(i,'-->',a)
```

```python
# looping with 2 collections
li=[1,2,3,4,5]
a=['a','b','c','d','e','f','g']
n=min(len(li),len(a))
for i in range(n):
    print(li[i],' ',a[i])
print()
# Pythonic Way
for i,a in zip(li,a):
    print(i,' ',a)
```

```python
li=[1,2,4,5,6,7,89,47]
for i in sorted(li):
    print(i,end=' ')
print()
for i in sorted(li,reverse=True):
    print(i,end=" ")
```

### List Comprehensions

```python
li= []
for i in range(10):
    a=i**2
    li.append(a)
print(li,end=' ')
print()
print(sum(li))
# Pythonic way
sum(i**2 for i in range(10))
```

```python
li=[]
for i in range(10):
    li.append(i**2)
li
```

```python
com_list=[i**2 for i in range(10)] # Comprehensive list
print(com_list)
```

```python
com_list=[ i**2 for i in range(10) if i%2==0]
print(com_list)
```

### Lambda Funtcion
 - Annonymous Function means that a function is without a name
 - Any lambda function in python will  be defined with lambda
 - Syntax : lambda arg : Expression

```python
def square(n):
    return n*n
square(6)
```

```python
a=lambda x:x*x
print(a(6))
```

```python
a=lambda x:x+2
print(a(10))
```

### Use of lambda with filter
    

```python
def filterlist(li):
    a=[]
    for i in li:
        if i%2 == 0:
            a.append(i)
    return a
li=[1,2,3,4,5,6,7,8]
filterlist(li)
```

```python
li = [1,2,3,4,5,6,7,8]
lam_li=list(filter(lambda x: (x%2==0),li))
print(lam_li)
```

```python
def squareList(li):
    a=[]
    for i in li:
        a.append(i**2)
    return a
li = [1,2,3,4,5,6,7,8]
squareList(li)
```

```python
li = [1,2,3,4,5,6,7,8]
map_list=list(map(lambda x:x**2,li))
print(map_list)
```

### lambda 

```python
def sumList(li):
    s=0
    for i in li:
        s+=i
    return s
li=[1,2,3,4,5,6,7,8]
sumList(li)
```

```python
from functools import reduce
li=[1,2,3,4,5,6,7,8]
s=reduce((lambda x,y:x+y),li)
print(s)
```
