### Visualizing Data
 - An intersting way to display DataFrame
 - With matplotlib

```python
import matplotlib.pyplot as plt
a1=[1,2,3,4]
a2=[5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-label') # To set name for x axix
plt.xlabel('y-label') # To set name for y axix
plt.show()
```

```python
marks = [79,82,75,63]
avg_marks = [56,72,78,89]
divisions = ['A','B','C','D']
index=np.arange(len(divisions))
plt.bar(divisions,marks,color='red',width=0.70)
```

```python
plt.barh(divisions,marks,color='red')
plt.show()
```

```python
lang=['Python','java','c','java']
rating=[75,45,55,65]
plt.pie(rating,labels=lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
